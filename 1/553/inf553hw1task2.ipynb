```python
def states(s):
    element = json.loads(s)
    return (element['state'], [element['business_id']])    
def businessToStars(s):
    element = json.loads(s)
    return (element['business_id'], element['stars'])
def businessToStarsOne(s):
    element = json.loads(s)
    return (element['business_id'], 1)

def state_with_business_average(s):
    element = json.loads(s)
    if element['business_id'] in average_map:
        return (element['state'], average_map[element['business_id']])
def state_with_business_average_mapValue(s):
    if s != None:
        return (s, 1)
def state_with_business_average_reduceByKey(a, b):
    if a != None and b != None:
        return (a[0] + b[0], a[1] + b[1])
def state_with_business_average_mapValue_div(a):
    if a != None:
        return a[0] / a[1]
```

```python
#average stars of each business
average_map = review \
        .map(lambda x : (json.loads(x)['business_id'], json.loads(x)['stars'])) \
        .mapValues(lambda x : (x , 1)) \
        .reduceByKey(lambda a, b : (a[0] + b[0], a[1] + b[1])) \
        .mapValues(lambda m : m[0] / m[1]) \
        .collectAsMap()
```

```python
global average_map
```

```python
#average star of each state
average_star_state = business \
                    .map(state_with_business_average).filter(lambda x : x is not None) \
                    .mapValues(lambda x : (x , 1)) \
                    .reduceByKey(lambda a, b : (a[0] + b[0], a[1] + b[1])) \
                    .mapValues(state_with_business_average_mapValue_div) \
                    .sortBy(lambda x : x[1], False)
```

```python
#m1 m2 runtime
average_m1_m2 = business \
                    .map(state_with_business_average).filter(lambda x : x is not None) \
                    .mapValues(lambda x : (x , 1)) \
                    .reduceByKey(lambda a, b : (a[0] + b[0], a[1] + b[1])) \
                    .mapValues(state_with_business_average_mapValue_div) \
                    .sortBy(lambda x : x[1], False)
```

```python
#m1
start_time = time.time()
m1_data = average_m1_m2.collect()
for x in range(5):
    print(m1_data[x][0])
print('{}s seconds'.format(time.time() - start_time))
```

```python
#m2
start_time = time.time()
m2_data = average_m1_m2.take(5)
for x in m2_data:
    print(x[0])
print('{}s seconds'.format(time.time() - start_time))
```
