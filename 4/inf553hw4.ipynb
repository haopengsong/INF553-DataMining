```python
from pyspark import SparkContext, SparkConf
import sys, time
```

```python
sc = SparkContext().getOrCreate()
```

```python
data = sc.textFile('/Users/haopengsong/Downloads/sample_data.csv')
```

```python
data.take(5)
```

```python
data_firstLine = data.take(1)
```

```python
data = data.filter(lambda x : x != data_firstLine[0])
```

```python
data.take(5)
```

```python
#check how many elements are in dataset
data.count()
```

```python
user_business = data.map(lambda x : x.split(",")) \
                    .map(lambda x : (x[0], [x[1]])) \
                    .reduceByKey(lambda a, b : a + b)
```

```python
user_business.take(1)
```

```python
#check if there are duplicate business elements 
user_business.map(lambda x : len(x[1])).sum()
```

```python
user_business = user_business.map(lambda x : (x[0], set(x[1])))
```

```python
#check if there are duplicate business elements 
user_business.map(lambda x : len(x[1])).sum()
```

```python
user_business.take(1)
```

```python
user_business_map = user_business.collectAsMap()
```

```python
def construct_graph(x):
    neighbor = set()
    same_review_num = 0
    for k,v in user_business_map.items():
        if k != x[0]:
            same_review_num = len(v.intersection(x[1]))
        if same_review_num >= 7:
            neighbor.add(k)
        
    return (x[0], neighbor)
s1 = time.time()
graph = user_business.map(construct_graph) \
                    .filter(lambda x : len(x[1]) != 0) \
                    .collectAsMap()
print('time: ' + str(time.time() - s1))
```
